```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data = np.random.power(2.,1000)
err  = 0.2*np.random.normal(0.,1.,1000) 
obs = data+err
```

```python
plt.plot(data,'g-',label='gaussian data')
plt.plot(obs,'r-.',label='observed data',lw=0.5)
plt.legend(loc='upper right')
plt.xlabel('time')
plt.ylabel('data')
#plt.xscale('log')
#plt.yscale('log')
plt.show()
plt.close()
# Se non chiudete l'istanza di finestra grafica con close, ognio plot finisce nella stessa finestra

plt.plot()
plt.plot(obs,'r--',label='observed data')
plt.legend()
plt.show()
plt.close()
```

```python
plt.hist(data,bins=20,histtype='step',color='green',label='data')
plt.hist(obs,bins=20,histtype='bar',color='red',label='obs',alpha=0.3)
plt.legend()
plt.show()
plt.close()
```

```python
data_h, bins_data = np.histogram(data, bins=20)
obs_h, bins_obs = np.histogram(obs, bins=20)

data_x = np.zeros(20)
obs_x = np.zeros(20)
for i in range(20):
    data_x[i] = 0.5*(bins_data[i] + bins_data[i+1])
    obs_x[i] = 0.5*(bins_obs[i] + bins_obs[i+1])
```

```python
plt.hist(data,bins=20,histtype='bar',color='green',label='data',alpha=0.3)
plt.plot(data_x, data_h, 'g-')
plt.hist(obs,bins=20,histtype='bar',color='red',label='obs',alpha=0.3)
plt.plot(obs_x, obs_h, 'r-')
plt.legend()
plt.show()
plt.close()
```

```python
plt.fill_between(data_x,data_h,obs_h,color='orange',alpha=0.5)
```

```python
def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

n = 100
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
plt.close()
```
